Ej 1-Removiendo sobresalientes

Usando los dos métodos elimina los sobresalientes del archivo datasets/outlierdat.csv

```python
import pandas as pd

Location = "datasets/outlierdata.csv"
df = pd.read_csv(Location)

q1 = df['Net Worth'].quantile(.25)
q3 = df['Net Worth'].quantile(.75)
iqr = q3-q1

meanNetWorth = df['Net Worth'].mean()

stdNetWorth = df['Net Worth'].std()

toprange = meanNetWorth + stdNetWorth * 1.96
botrange = meanNetWorth - stdNetWorth * 1.96

copydf = df
copydf = copydf.drop(copydf[copydf['Net Worth'] > toprange].index)
copydf = copydf.drop(copydf[copydf['Net Worth'] < botrange].index)

copydf
```

Ej-10- Convirtiendo categorías Strings a numéricas

Convertir las categorías desde "freshman" hasta "senior" con los numerales 1 al 4 para el dataset: datasets/gradesdatawithyear.csv

```python
import pandas as pd

Location = "datasets/gradesdatawithyear.csv"
df = pd.read_csv(Location)

df
```

```python
def score_to_numeric(x):
    if x=='junior':
        return 1
    if x=='sophomore':
        return 2
    if x=='freshman':
        return 3
    if x=='senior':
        return 4
```

```python
df['sophomore'].apply(score_to_numeric)
```
